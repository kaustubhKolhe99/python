```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df.head()
```

# getting rid of null values

```python
median_bedroom = df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedroom) 	# replaces null values in bedroom column
```

```python
df.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.predict([[3000, 3, 40]])
```

```python
reg.coef_[0] * 3000 + reg.coef_[1] * 3 +reg.coef_[2] * 40 + reg.intercept_	
```
