```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names	# corresponding to 0, 1, 2 in target in above df
```

```python
df[df.target ==1].head()
```

```python
df['flower name'] = df.target.apply(lambda t: iris.target_names[t])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df2.head()
```

```python
plt.xlabel('selal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', marker='+')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='red', marker='+')
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', marker='+')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='red', marker='+')
```

```python
df.head()
```

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['target', 'flower name'], axis=1), df.target, test_size=0.2)
```

```python
from sklearn.svm import SVC
svm = SVC(C=10)	# c -> regularization
svm.fit(X_train, y_train)
```

```python
svm.score(X_test, y_test)
```
